```python
# write your code here
# 1. Download the dataset and read it into a pandas dataframe:!
df_sfcrime = pd.read_csv('https://cocl.us/sanfran_crime_dataset/Police_Department_Incidents_-_Previous_Year__2016_.csv')
print('Dataset downloaded and read into a pandas dataframe!')
```

```python
# Rename the column and Set - index column
#df_survey.columns = ['', 'Very interested', 'Somewhat interested','Not interested']
#df_survey.set_index('',inplace = True)
#df_sfcrime.dtypes
#df_sfcrime['PdDistrict'].value_counts()    
#df_sfcrime[(df_sfcrime['Category']!= "NON-CRIMINAL") & (df_sfcrime['Category']!= "OTHER OFFENSES")].groupby('PdDistrict',axis=0).count()
#df_survey['Total'] = df_survey.sum(axis=1)
#df_survey.drop(['Total'], axis=1, inplace=True)
df_sftot_crime = df_sfcrime.groupby('PdDistrict',axis=0).count()[['IncidntNum']]
df_sftot_crime.sort_values(by='IncidntNum',ascending=False,inplace=True)
df_sftot_crime.reset_index(inplace=True)
df_sftot_crime.columns = ['Neighborhood','Count']
df_sftot_crime
```

```python
#world_geo = r'world_countries.json' # geojson file
sf_geo = r'san-francisco.geojson' # geojson file
# create a plain world map
sf_map = folium.Map(location=[37.773972, -122.431297], zoom_start=12)
```

```python
# generate choropleth map using the total immigration of each country to Canada from 1980 to 2013
sf_map.choropleth(
    geo_data=sf_geo,
    data=df_sftot_crime,
    columns=['Neighborhood','Count'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Crime Rate in San Francisco',
    reset=True
)

# display map
sf_map
```
